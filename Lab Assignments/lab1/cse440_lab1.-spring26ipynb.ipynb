```python
from nltk.corpus import stopwords
movie_reviews.fileids()
#movie_reviews.categories()
```

```python

t3=movie_reviews.raw("neg/cv965_26688.txt")
words=word_tokenize(t3.lower())
stop_words=stopwords.words("english")

filtered_words=[]

for i in words:
    if i not in stop_words and i.isalpha():
        filtered_words+=[i]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words=[]
for i in filtered_words:
    lemmatized_words+=[lemmatizer.lemmatize(i,pos='v')]
word_count=FreqDist(lemmatized_words)
```

```python
words=[]
counts=[]

ontop=word_count.most_common(20)

for itm in ontop:
    word= itm[0]
    freq= itm[1]
    words.append(word)
    counts.append(freq)

plt.figure(figsize=(13,4))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.xlabel("Top 20 Words")
plt.ylabel("Frequency")
plt.show()
```

```python
import string
from nltk import FreqDist
from nltk.util import ngrams
reuters.fileids()
```

```python
import numpy as np
txt1= reuters.raw('test/16512').lower()

w_tokens= word_tokenize(txt1)
stop_words = stopwords.words("english")

filtered_words= []

for w in w_tokens:
    if w.isalpha():
        if w not in stop_words:
            filtered_words.append(w)

freq1= FreqDist(filtered_words)
print("Top 12 frequent words from the text:")

max= freq1.most_common(12)

top12= []
for itm in max:
    top12.append(itm[0])
print(top12)
print()

bigrams=list(ngrams(filtered_words,2))

size=len(top12)
m=np.zeros((size, size), dtype=int)

for x in bigrams:
    w1 = x[0]
    w2 = x[1]

    if w1 in top12 and w2 in top12:
        i=top12.index(w1)
        j=top12.index(w2)
        m[i][j]=m[i][j]+1
print("The co occurrence matrix: ")
print(m)
```
